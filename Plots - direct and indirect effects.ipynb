```python
import plots
from config import german, compas, adult, tags
import config
import pickle
```

```python
parameter_dicts = {}
for term_names in config.regularization_terms:
    parameter_name = '_'.join(term_names)
    dataset_dicts = []
    for dataset_info in [german, compas, adult]:
        evaluations_dict = pickle.load(open("evaluations/" + dataset_info['tag'] + "_" + config.parameters['name'] + "_" + parameter_name + ".p", "rb"))
        averaged_testing_dicts = [d['averaged'] for d in evaluations_dict['testing']]
        dataset_dicts.append((dataset_info['name'], dataset_info['tag'], averaged_testing_dicts))
    parameter_dicts[parameter_name] = dataset_dicts
```

```python
effects_to_plot = [
    ('IND', 'N-IND'),
    ('SEP', 'N-SEP'),
    ('SUF', 'N-SUF'),
    ('IND', 'N-SEP'),
    ('SEP', 'N-IND'),
    ('SEP', 'N-SUF'),
    ('SUF', 'N-SEP'),
    ('IND', 'N-SUF'),
    ('SUF', 'N-IND'),
    ('BAL', 'N-SEP'),
    ('BAL', 'N-SUF'),
    ('-ACC', 'N-SEP'),
    ('-ACC', 'N-SUF'),
    ('IND', '-ACC'),
    ('SEP', '-ACC'),
    ('SUF', '-ACC'),
    ('BAL', '-ACC'),
    ('-ACC', '-ACC'),
    ('IND', 'BAL'),
    ('SEP', 'BAL'),
    ('SUF', 'BAL'),
    ('BAL', 'BAL'),
    ('-ACC', 'BAL'),
    ('BAL', 'N-IND'),
    ('-ACC', 'N-IND'),
]
```

```python
def plot_effects(of, on):
    plots.plot_effect_on(parameter_dicts[tags[of]['parameter']], of, on)
```

```python
for of, on in effects_to_plot:
    plot_effects(of, on)
```
