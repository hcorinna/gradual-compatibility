```python
import data
from optimize_lambda import CustomCrossValidator
import utils
import config
import warnings
warnings.filterwarnings('ignore')
```

```python
def find_optimal_lambda(dataset_info):
    X, y = data.get_X_y(dataset_info)

    cross_validator = CustomCrossValidator(X, y)
    l2_lambda = cross_validator.cross_validate(config.lambdas, num_folds=config.number_of_lambda_folds)

    utils.save_lambda(dataset_info, l2_lambda)
```

```python
for dataset_info in config.datasets:
    print('\n\n' + dataset_info['name'])
    find_optimal_lambda(dataset_info)
```
